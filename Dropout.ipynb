# Dropout
Dropout [1] is a technique for regularizing neural networks by randomly setting some output activations to zero during the forward pass. In this exercise, you will implement a dropout layer and modify your fully connected network to optionally use dropout.

[1] [Geoffrey E. Hinton et al, "Improving neural networks by preventing co-adaptation of feature detectors", arXiv 2012](https://arxiv.org/abs/1207.0580)

## Inline Question 1:
What happens if we do not divide the values being passed through inverse dropout by `p` in the dropout layer? Why does that happen?

## Answer:
### 1. The Impact of Failing to Scale by $1/p$
If we do not divide the activations by the keep probability $p$ during training, the model will produce **incorrect predictions at test time**. This occurs because the scale of the activations arriving at the next layer during testing will be significantly higher than the scale the network "learned" to handle during training.

### 2. Expected Output without Scaling
During training, a dropout layer randomly zeros out neurons. Without scaling, the **expected value** $E[\hat{y}]$ of a neuron producing output $\hat{y}$ is:

$$E[\hat{y}] = p\hat{y} + (1 - p)0 = p\hat{y}$$

* **$p\hat{y}$**: The probability of the neuron being active.
* **$(1 - p)0$**: The probability of the neuron being dropped (outputting zero).

Since $E[\hat{y}] = p\hat{y}$, the outputs during training are "dimmed" by a factor of $p$. However, at test time, all neurons are active ($p=1$), meaning the outputs will be roughly $1/p$ times larger than what the subsequent layers expected based on training.


---

### 3. The Advantage of Inverse Dropout
By dividing by $p$ (or multiplying by $1/p$) during the training phase, we normalize the expected value so that it matches the test-time scale:

$$E[\hat{y}] = \frac{p}{p}\hat{y} + \frac{(1 - p)}{p}0 = \hat{y}$$

**Conclusion:** Scaling by $1/p$ during training ensures that the magnitude of the signal remains consistent between training and testing. This allows us to keep the test-time inference simple and efficient, as we do not need to perform any additional scaling or modifications when the model is deployed.

# Regularization Experiment
As an experiment, we will train a pair of two-layer networks on 500 training examples: one will use no dropout, and one will use a keep probability of 0.25. We will then visualize the training and validation accuracies of the two networks over time.

```python
# Train two identical nets, one with dropout and one without.
np.random.seed(231)
num_train = 500
small_data = {
    'X_train': data['X_train'][:num_train],
    'y_train': data['y_train'][:num_train],
    'X_val': data['X_val'],
    'y_val': data['y_val'],
}

solvers = {}
dropout_choices = [1, 0.25]
for dropout_keep_ratio in dropout_choices:
    model = FullyConnectedNet(
        [500],
        dropout_keep_ratio=dropout_keep_ratio
    )
    print(dropout_keep_ratio)

    solver = Solver(
        model,
        small_data,
        num_epochs=25,
        batch_size=100,
        update_rule='adam',
        optim_config={'learning_rate': 5e-4,},
        verbose=True,
        print_every=100
    )
    solver.train()
    solvers[dropout_keep_ratio] = solver
    print()
```

```python
# Plot train and validation accuracies of the two models.
train_accs = []
val_accs = []
for dropout_keep_ratio in dropout_choices:
    solver = solvers[dropout_keep_ratio]
    train_accs.append(solver.train_acc_history[-1])
    val_accs.append(solver.val_acc_history[-1])

plt.subplot(3, 1, 1)
for dropout_keep_ratio in dropout_choices:
    plt.plot(
        solvers[dropout_keep_ratio].train_acc_history, 'o', label='%.2f dropout_keep_ratio' % dropout_keep_ratio)
plt.title('Train accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(ncol=2, loc='lower right')

plt.subplot(3, 1, 2)
for dropout_keep_ratio in dropout_choices:
    plt.plot(
        solvers[dropout_keep_ratio].val_acc_history, 'o', label='%.2f dropout_keep_ratio' % dropout_keep_ratio)
plt.title('Val accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(ncol=2, loc='lower right')

plt.gcf().set_size_inches(15, 15)
plt.show()
```

## Inline Question 2:
Compare the validation and training accuracies with and without dropout -- what do your results suggest about dropout as a regularizer?

## Answer:
### 1. Training Performance: The Cost of Regularization
During the training phase, the model **without dropout** typically achieves higher accuracy.
Without dropout, the network is unrestricted; it can utilize every available neuron to memorize the specific features and noise within the training set.
While the accuracy is higher, this often indicates that the model is **overfitting** to the training data rather than learning generalized patterns.

---

### 2. Validation Performance: Improved Generalization
During the validation phase, the model **with dropout** generally produces superior results.
Dropout acts as a powerful **regularizer** by randomly "dropping" neurons during each forward pass. This prevents the network from developing complex co-adaptations where it relies too heavily on specific, individual neurons.
Because no single neuron can be relied upon, the model is forced to learn more **robust, redundant features** that are distributed across the entire architecture.



---

### 3. Conclusion: Dropout's Role in Robustness
The results suggest that dropout is an effective tool for controlling overfitting. Although it may slightly decrease training performance by making the optimization task more difficult, it significantly enhances the model's ability to **generalize** to unseen data. By introducing stochastic noise during training, dropout ensures the network remains robust and prevents it from simply memorizing the training samples.
